```python
def validate_age(age):
    errors = []
    if not isinstance(age, int):
        return False, errors
    if age < 0:
        errors.append('Age cannot be negativve')
    if age > 120:
        errors.append('Age too high (max 120)')
    is_valid = len(errors) == 0
    return is_valid, errors

valid, errs = validate_age(25)
print(valid)
valid, errs = validate_age(-5)
print(valid, errs)
```

```python
ages = [25, -5, 30, 150, 45, 'invalid', 67, 0, 120]
valid_ages = []
invalid_ages = []
for age in ages:

    if not isinstance(age, (int, float)):
        invalid_ages.append((age, 'Not a nummber'))
        continue

    if 0 <= age <= 120:
        valid_ages.append(age)
    else:
        invalid_ages.append((age, 'Out of range'))
print(f'Valid: {len(valid_ages)} items')
print(f'Invalid: {len(invalid_ages)} items')
print(f'Invalid data: {invalid_ages}')
```

```python
def calculate_average(scores):
    if scores is None:
        return None, 'Error: No scores provided'
    if len(scores) == 0:
        return None, 'Error: Empty score list'

    valid_scores = [s for s in scores if 0 <= s <= 100]
    if len(valid_scores) == 0:
        return None, 'Error: No valid scores'

    avg = sum(valid_scores) / len(valid_scores)
    return avg, 'Success'

result, msg = calculate_average([95, 87, 92])
print(f'{msg}: {result}')
```
