```python
#*args EX Sum_Any_Numbers
#function accepts any number of arguments
def add_all(*args):
    print(f'Received: {args}') #Tuple
    return sum(args)
```

```python
#call with different amounts
print(add_all(1, 2))
```

```python
print(add_all(1, 2, 3, 4, 5))
```

```python
print(add_all(10, 20, 30, 40, 50, 60))
```

```python
#processing *args
# *args is tuple - loop through it
def find_max(*args):
    if not args:
        return None
    max_val = args[0]
    for num in args:
        if num > max_val:
            max_val = num
    return max_val
print(find_max(5, 2, 9, 1, 7))
```

```python
print(find_max(100, 50, 75))
```

```python
# or just use max():
def find_max(*args):
    return max(args) if args else None
```

```python
# **kwargs example - flexible config
#function accepts any keyword arguments
def create_profile(**kwargs):
    print('Profile created:')
    for key, value in kwargs.items():
        print(f' {key}: {value}')
```

```python
#call with different fields
create_profile(name='Alice', age=25)
```

```python
create_profile(name='Bob', age=30, city='NYC', job='Engineer')
```

```python
#Combined example
#all three types of parameters
def process_data(operation, *args, **kwargs):
    print(f'Operation: {operation}')
    print(f'Data: {args}')
    print(f'Options: {kwargs}')
```

```python
#call with all types
process_data(
    'analyze',     # regular
    10, 20, 30,    # *args
    method='mean',   # **kwargs
    precision=2
)
```
