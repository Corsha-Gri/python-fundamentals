```python
#define function with parameter 
def greet(name):
    print(f'Hello, {name}!')
    print('Welcome to Python!')
```

```python
#call with different arguments
greet('Alice')
```

```python
greet('Bob')
```

```python
greet('Carol')
```

```python
#multiple parameters
#function with two parameters
def introduce(name, age):
    print(f'My name is {name}')
    print(f'I am {age} years old')
```

```python
#call with two arguments
introduce('Alice', 25)
```

```python
#function that returns value
def add(a, b):
    result = a + b
    return result
```

```python
#call and store result
sum1 = add(5, 3)
print(sum1)
```

```python
#use returned value in calculation
total = add(5, 3) + add(10, 2)
print(total)
```

```python
#return vs print
#function that PRINTS
def add_print(a, b):
    result = a + b
    print(result)
```

```python
#function that RETURNS
def add_return(a, b):
    result = a + b
    return result
```

```python
#try to use them:
x = add_print(5, 3)
```

```python
print(x)
```

```python
y = add_return(5, 3)
```

```python
print(y)
```

```python
#docstring 
def calculate_average(numbers):
    """Calculate the average of a list of numbers.
    Args:
        numbers: List of numeric values

    Returns:
        Float representing the average
    Example:
        >>> calculate_average([10, 20, 30])
        20.0
    """
    return sum(numbers) / len(numbers)
```

```python
#access docstring
print(calculate_average.__doc__)
```
