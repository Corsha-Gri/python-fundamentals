```python
#best practicce: use with
with open('sample.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
#file automatically closed here!
#Even if exception occured
```

```python
#can still use content after 'with' block
print('\nLength:', len(content))
```

```python
#this is the STANDARD way
#Use this pattern always!
#with = try-finally behind the scenes
#Guarantees file.close() gets called
```

```python
#Method 1: read() - whole file
with open('sample.txt', 'r') as file:
    content = file.read()
    print(type(content))      #Str
```

```python
# Method 2: readlines() - list of lines
with open('sample.txt', 'r') as file:
    lines = file.readlines()
    print(type(lines))    #list
    print(lines[0])      #First line
```

```python
#Method 3: loop(best for large files)
with open('sample.txt', 'r') as file:
    for line in file:
        print(line.strip())   # Remove \n
```

```python
#loop doesnt load entire file into memory
# efficient for huge files
```

```python
#writing files
#write mode - creates new file or overwrites
with open('output.txt', 'w') as file:
    file.write('Hello, World!\n')
    file.write('This is line 2\n')
    file.write('This is line 3\n')
    #file now contains 3 lines
```

```python
#Append mode - adds to end
with open('output.txt', 'a') as file:
    file.write('This is line 4\n')
```

```python
#File now 4 lines
#write doesnt add newlines automatically
#must inclde \n yourself
```

```python
#read it back to verify
with open('output.txt', 'r') as file:
    print(file.read())
```

```python
#writing list data
#student scores to save
scores = [95, 87, 92, 78, 88, 94]
```

```python
#method 1: Loop
with open('scores.txt', 'w') as file:
    for score in scores:
        file.write(f'{score}\n')
```

```python
#method 2: writelines with comprehension
with open('scores.txt', 'r') as file:
    loaded = [int(line.strip()) for line in file]
print(loaded)
```

```python
#Wrting CSV Files
import csv
```

```python
#data to save
students = [
    ['Name', 'Age', 'Score'],  #Header
    ['Alice', 25, 95],
    ['Bob', 30, 87],
    ['Carol', 28, 92]
]
```

```python
#Write CSV file
with open('students.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(students)
```

```python
#Creates:
#name, age, score
#newline='' prevents extra blank lines
```
