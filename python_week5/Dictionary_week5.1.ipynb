```python
#Looping Examples
grades = {'math': 85, 'english': 92, 'science': 88}
```

```python
#loop through keys only
for subject in grades:
    print(subject)
```

```python
#loop through values only
for grade in grades.values():
    print(grade)
```

```python
#loop through both
for subject, grade in grades.items():
    print(f'{subject}: {grade}')
```

```python
#use .items() most of the time!
```

```python
#calculate stats from dictionary
prices = {
    'laptop': 999.99,
    'mouse': 25.50,
    'keyboard': 79.99,
    'monitor': 299.99
}
```

```python
#Total value
total = sum(prices.values())
print(f'Total: ${total:.2f}')
```

```python
#average price
avg = total / len(prices)
print(f'Average: ${avg:.2f}')
```

```python
#most expensive
max_item = max(prices, key=prices.get)
print(f'Most expensive: {max_item}')
```

```python
student = {'name': 'Alice', 'age':25}
```

```python
if 'gpa' in student:
    print(student['gpa'])
else:
    print('GPA not recorded')
```

```python
#Or use .get() with default
gpa = student.get('gpa', 0.0)
print(f'GPA: {gpa}')
```

```python
#create dict of squares
squares = {x: x**2 for x in range(1,6)}
print(squares)
```

```python
#double all values
prices = {'apple': 1.50, 'banana': 0.75}
doubled = {k: v*2 for k, v in prices.items()}
print(doubled)
```

```python
#filter high grades only
grades = {'math': 85, 'english': 92, 'science': 78}
high = {k: v for k, v in grades.items() if v >= 80}
print(high)
```
