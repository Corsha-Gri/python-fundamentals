```python
grades = [
[95, 87, 92],
[88, 90, 85],
[92, 88, 94],
[78, 85, 80]
]
```

```python
for student_grades in grades:
    avg = sum(student_grades) / len(student_grades)
```

```python
print(f'Average: {avg:.1f}')
```

```python
avgs = [sum(row)/len(row) for row in grades]
```
