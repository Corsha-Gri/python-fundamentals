```python
responses = ["yes", "no", "maybe", "42", "banana", "yesss"]
```

```python
print('Invalid answers (with line numbers):')
for idx, answer in enumerate(responses, start=1):
    if answer.lower() not in ['yes', 'no']:
        print(f' Line (idx:2d): "{answer}" <- not allowed')
```

```python
sales = [120.50, 45.00, 890.75, 15.20, 320.00, 675.30]
```

```python
total_sales = 0
```

```python
high_value_count = 0
```

```python
for amount in sales:
    total_sales += amount
    if amount > 500:
        high_value_count += 1
```

```python
average_sale = total_sales / len(sales)
```

```python
print(f'Total sales: ${total_sales:.2f}')
```

```python
print(f'Average sale: ${average_sale:,.2f}')
```

```python
print(f'High-value sales (>500); {high_value_count}')
```

```python
import datetime
```

```python
dates = [
    datetime.date(2025, 10, 13),
    datetime.date(2025, 10, 14),
    datetime.date(2025, 10, 15),
    datetime.date(2025, 10, 16),
    datetime.date(2025, 10, 17),
    datetime.date(2025, 10, 18),
    datetime.date(2025, 10, 19),
    datetime.date(2025, 10, 20),
]
```

```python
weekdays = []
```

```python
for dt in dates:
    if dt.weekday() < 5:
        weekdays.append(dt)
```

```python
print("weekdays only:")
for d in weekdays:
    print(d.strftime("%A, %Y-%m-%d:"))
```
